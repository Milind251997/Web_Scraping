```python
url = "https://www.amazon.com/s?k=Wallpaper&rh=n%3A2242314011&ref=nb_sb_noss"
page = requests.get(url)
print("Legality Response number from our URL is:", page) # to show the response output from the webpage
soup = BeautifulSoup(page.content)

driver =webdriver.Chrome(r"C:\Users\Admin\Desktop\Web Scrapping\chromedriver.exe")
driver.get(url)
driver.maximize_window()
```

```python
# getting all the urls to open each page from our above search results
urls =[]
for i in range(0,4): # upto 4 pages
    for x in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(x.get_attribute('href'))
    driver.find_element_by_xpath("//li[@class='a-last']/a").click()
    time.sleep(3)
```

```python
# all the initial empty lists
brand=[]
prod_name=[]
rating=[]
price= []
prod_url=[]
ot_det = []

# obtaining all the details one after another
for i in urls[:70]:
    driver.get(i)
    time.sleep(3)
    
    # getting the brand name
    try:
        brand_name=driver.find_element_by_xpath('//*[@id="productDetails_techSpec_section_1"]/tbody/tr[1]/td')
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append('-')    
                
      # getting the product name
    try:
        product=driver.find_element_by_xpath('//h1[@id="title"]')
        prod_name.append(product.text)
    except NoSuchElementException:
        prod_name.append('-')         
        
      # getting the rating
    try:
        ratestar=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
        rating.append(ratestar.text)
    except NoSuchElementException:
        rating.append("No Rating")       
                  
                
      # getting the price
    try:
        prod_price=driver.find_element_by_id('//*[@id="price_inside_buybox"]')
        price.append(prod_price.text)
    except NoSuchElementException:
        price.append('-')           
                           
     # getting the product url
    try:
        pro_url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
        for i in pro_url:
            prod_url.append(i.get_attribute("href"))
    except NoSuchElementException:
        prod_url.append('-')
```

```python
# obtaining all the details one after another
for i in urls[70:]:
    driver.get(i)
    time.sleep(3)
    
    # getting the brand name
    try:
        brand_name=driver.find_element_by_xpath('//*[@id="productDetails_techSpec_section_1"]/tbody/tr[1]/td')
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append('-')    
                
      # getting the product name
    try:
        product=driver.find_element_by_xpath('//h1[@id="title"]')
        prod_name.append(product.text)
    except NoSuchElementException:
        prod_name.append('-')         
        
      # getting the rating
    try:
        ratestar=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
        rating.append(ratestar.text)
    except NoSuchElementException:
        rating.append("No Rating")       
                  
                
      # getting the price
    try:
        prod_price=driver.find_element_by_id('//*[@id="price_inside_buybox"]')
        price.append(prod_price.text)
    except NoSuchElementException:
        price.append('-')           
                           
     # getting the product url
    try:
        pro_url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
        for i in pro_url:
            prod_url.append(i.get_attribute("href"))
    except NoSuchElementException:
        prod_url.append('-')
```

```python
# creating the entire dataframe now
item_df = pd.DataFrame({})
item_df['Brand Name'] = brand
item_df['Product Name'] = prod_name
item_df['Rating'] = rating
item_df['Price'] = price
item_df['Product URLs'] = prod_url  
item_df
```

```python
item_df.to_csv('Wallpaper_Data.csv', index = True)
```
